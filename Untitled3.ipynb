```python
with open("parallel_image_processing/mpi/mpi_cpu_parallel.py", "w") as f:
    f.write('''\
from mpi4py import MPI
import cv2
import numpy as np
import time

comm = MPI.COMM_WORLD
rank = comm.Get_rank()
size = comm.Get_size()

IMG_PATH = "/content/image2 (1).jpg"
OUTPUT_PATH = "mpi_cpu_output.jpg"

if rank == 0:
    img = cv2.imread(IMG_PATH)
    if img is None:
        print(f"Error: Image not found at {IMG_PATH}. Please ensure it is uploaded.")
        comm.Abort(1)

    # Split the image into chunks for each process
    h, w, _ = img.shape
    chunks = np.array_split(img, size, axis=0)
else:
    chunks = None

# Scatter chunks to processes
chunk = comm.scatter(chunks, root=0)

# Convert chunk to greyscale
gray_chunk = cv2.cvtColor(chunk, cv2.COLOR_BGR2GRAY)

# Gather processed chunks
gathered = comm.gather(gray_chunk, root=0)

if rank == 0:
    # Merge chunks and save the final output
    gray_image = np.vstack(gathered)
    cv2.imwrite(OUTPUT_PATH, gray_image)
    print("RESULT: MPI_CPU Completed.")
    ''')
```

```python
with open("parallel_image_processing/mpi/cuda_kernel.py", "w") as f:
    f.write('''\
from numba import cuda
import numpy as np

@cuda.jit
def cuda_bgr_to_grayscale(bgr_image, gray_image):
    # Map thread ID
    row, col = cuda.grid(2)
    if row < gray_image.shape[0] and col < gray_image.shape[1]:
        r = bgr_image[row, col, 2]
        g = bgr_image[row, col, 1]
        b = bgr_image[row, col, 0]
        gray_image[row, col] = 0.299 * r + 0.587 * g + 0.114 * b
    ''')
```

```python
with open("parallel_image_processing/mpi/mpi_cuda_hybrid.py", "w") as f:
    f.write('''\
from mpi4py import MPI
import cv2
import numpy as np
from numba import cuda
import time
from cuda_kernel import cuda_bgr_to_grayscale

comm = MPI.COMM_WORLD
rank = comm.Get_rank()
size = comm.Get_size()

IMG_PATH = "/content/image2 (1).jpg"
OUTPUT_PATH = "mpi_cuda_output.jpg"

if rank == 0:
    img = cv2.imread(IMG_PATH)
    if img is None:
        print(f"Error: Image not found at {IMG_PATH}. Please ensure it is uploaded.")
        comm.Abort(1)

    # Split the image into chunks for each process
    h, w, _ = img.shape
    chunks = np.array_split(img, size, axis=0)
else:
    chunks = None

# Scatter chunks to processes
chunk = comm.scatter(chunks, root=0)

# Allocate device memory for bgr and gray images
bgr_device = cuda.to_device(chunk)
gray_device = cuda.device_array((chunk.shape[0], chunk.shape[1]), dtype=np.uint8)

# Launch CUDA kernel
threads_per_block = (16, 16)
blocks_per_grid = (int(np.ceil(chunk.shape[0] / threads_per_block[0])),
                   int(np.ceil(chunk.shape[1] / threads_per_block[1])))
cuda_bgr_to_grayscale[blocks_per_grid, threads_per_block](bgr_device, gray_device)

# Copy back to host
gray_chunk = gray_device.copy_to_host()

# Gather processed chunks
gathered = comm.gather(gray_chunk, root=0)

if rank == 0:
    # Merge chunks and save the final output
    gray_image = np.vstack(gathered)
    cv2.imwrite(OUTPUT_PATH, gray_image)
    print("RESULT: MPI_CUDA Completed.")
    ''')
```

```python
import cv2
from matplotlib import pyplot as plt

# Paths to the images
ORIGINAL_IMAGE_PATH = "image2 (1).jpg"  # Replace with the original image file name
GRAYSCALE_IMAGE_PATH = "mpi_cpu_output.jpg"  # Processed grayscale image

# Load the images
original_image = cv2.imread(ORIGINAL_IMAGE_PATH)
gray_image = cv2.imread(GRAYSCALE_IMAGE_PATH, cv2.IMREAD_GRAYSCALE)

# Check if both images exist
if original_image is not None and gray_image is not None:
    plt.figure(figsize=(12, 6))

    # Display the original image
    plt.subplot(1, 2, 1)
    plt.imshow(cv2.cvtColor(original_image, cv2.COLOR_BGR2RGB))  # Convert BGR to RGB for proper colors
    plt.axis('off')
    plt.title("Original Image")

    # Display the processed grayscale image
    plt.subplot(1, 2, 2)
    plt.imshow(gray_image, cmap='gray')  # Display in grayscale mode
    plt.axis('off')
    plt.title("Grayscale Image (MPI-CUDA)")

    plt.tight_layout()
    plt.show()
else:
    print("Error: Could not load one or more images.")
    if original_image is None:
        print(f"Original image not found at {ORIGINAL_IMAGE_PATH}")
    if gray_image is None:
        print(f"Grayscale image not found at {GRAYSCALE_IMAGE_PATH}")
```
