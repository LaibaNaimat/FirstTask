```python
%%writefile parallel_image_processing/mpi/mpi_grayscale.py
from mpi4py import MPI
import cv2
import numpy as np

comm = MPI.COMM_WORLD
rank = comm.Get_rank()
size = comm.Get_size()

# ---------- MASTER PROCESS ----------
if rank == 0:
    img = cv2.imread("/content/parallel_image_processing/testtt.jpg")

    if img is None:
        raise ValueError("Image not found. Check path!")

    height = img.shape[0]

    # Split image rows for each process
    chunks = np.array_split(img, size, axis=0)

else:
    chunks = None

# ---------- DISTRIBUTE IMAGE ----------
chunk = comm.scatter(chunks, root=0)

# ---------- PARALLEL PROCESSING ----------
gray_chunk = cv2.cvtColor(chunk, cv2.COLOR_BGR2GRAY)

# ---------- GATHER RESULTS ----------
gathered = comm.gather(gray_chunk, root=0)

# ---------- SAVE FINAL IMAGE ----------
if rank == 0:
    final_image = np.vstack(gathered)
    cv2.imwrite("mpi_output.jpg", final_image)
    print("MPI Grayscale image saved as mpi_output.jpg")
```
